```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as numpy
import seaborn as sns
```

```python
df = pd.read_csv("supermarket.csv")
```

```python
# Country-wise total sales
country_sales = df.groupby("Country")["Sales"].sum()

# Highest & lowest country
highest = country_sales.idxmax()
lowest = country_sales.idxmin()

# Data for chart
countries = [highest, lowest]
sales = [country_sales[highest], country_sales[lowest]]

# Bar chart
plt.figure(figsize=(7,5))
plt.bar(countries, sales)
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.title("Highest vs Lowest Sales by Country")
plt.show()
```

```python
# Region-wise total sales
region_sales = df.groupby("Region")["Sales"].sum()

# Highest sales region
top_region = region_sales.idxmax()

# Plot
plt.figure(figsize=(7,5))
plt.bar(region_sales.index, region_sales.values)
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.title("Region-wise Sales Comparison")
plt.show()
```

```python
# State-wise total sales
state_sales = df.groupby("State")["Sales"].sum()

# Highest sales state
top_state = state_sales.idxmax()

# Plot
plt.figure(figsize=(8,5))
plt.bar(state_sales.index, state_sales.values)
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.title("State-wise Sales Comparison")
plt.xticks(rotation=45)
plt.show()
```

```python
# Category-wise total sales
category_sales = df.groupby("Category")["Sales"].sum()

# Plot
plt.figure(figsize=(7,5))
plt.bar(category_sales.index, category_sales.values)
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.title("Category-wise Sales")
plt.show()
```

```python
# Sub-Category-wise total sales
sub_category_sales = df.groupby("Sub-Category")["Sales"].sum()

# Plot
plt.figure(figsize=(10,5))
plt.bar(sub_category_sales.index, sub_category_sales.values)
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.title("Sub-Category-wise Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
segment_sales = df.groupby("Segment")["Sales"].sum()

plt.figure(figsize=(6,4))
plt.bar(segment_sales.index, segment_sales.values, color="skyblue")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")
plt.title("Segment-wise Sales")
plt.show()
```

```python
ship_sales = df.groupby("Ship Mode")["Sales"].sum()

plt.figure(figsize=(6,4))
plt.bar(ship_sales.index, ship_sales.values, color="orange")
plt.xlabel("Ship Mode")
plt.ylabel("Total Sales")
plt.title("Sales by Shipping Mode")
plt.show()
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,4))
plt.barh(top_products.index, top_products.values, color="green")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.title("Top 5 Products by Sales")
plt.show()

```

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(6,6))
plt.pie(region_sales.values, labels=region_sales.index, autopct="%1.1f%%", startangle=140)
plt.title("Region-wise Sales Share")
plt.show()
```
